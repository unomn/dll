```python
#计算分子指纹
from rdkit import Chem
from rdkit.Chem import AllChem
import pandas as pd

# 导入文件读取 SMILES 数据
smiles_df = pd.read_csv("smiles.csv")
smiles_list = smiles_df["SMILES"].dropna().tolist()  # 去除空值

# 转换为分子对象并计算 Morgan 指纹
mols = [Chem.MolFromSmiles(smile) for smile in smiles_list if Chem.MolFromSmiles(smile)]
mols_fps = [AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=1024) for mol in mols]

# 将指纹转换为 DataFrame 并导出为 CSV
fingerprint_matrix = [[int(bit) for bit in fp.ToBitString()] for fp in mols_fps]
fingerprint_df = pd.DataFrame(fingerprint_matrix, columns=[f'Bit_{i}' for i in range(1024)])
fingerprint_df.to_csv("molecule_fingerprints1.csv", index=False)

print("Fingerprints have been successfully exported to 'molecule_fingerprints1.csv'.")
```
